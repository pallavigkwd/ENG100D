# Viasat Specific Data

### Energy Consumption (kWh) of Meter 6700126 at Viasat from 2021-2022 (15 min intervals)

```python
# https://docs.google.com/spreadsheets/d/1ezVDSIUbolbGZscLJ-8Ve8jMdory6p7O/edit#gid=1480136096 
```

```python
viasatConsumption = pd.read_csv('data/viasatConsumption.csv')
viasatConsumption
```

```python

# # avg consumption at all times of day (15 min intervals)
time_groups = viasatConsumption.groupby('Start Time')
time_totals = []
for start_time, group in time_groups:
    avg_consumption = group['Consumption'].mean()
    time_totals.append((start_time, avg_consumption))
    # print(f'Average consumption at {start_time}: {avg_consumption:.2f}')
    
print('\nMax Consumption Time (avg): ', max(time_totals, key=lambda x: x[1]))
print('\nMin Consumption Time (avg): ', min(time_totals, key=lambda x: x[1]))

# total consumption per day
date_groups = viasatConsumption.groupby('Date')
date_totals = []
for date, group in date_groups:
    total_consumption = group['Consumption'].sum()
    date_totals.append((date, total_consumption))
    # print(f'Total consumption on {date}: {total_consumption:.2f}')
    
print('\nMax Consumption Date: ', max(date_totals, key=lambda x: x[1]))
print('\nMin Consumption Date: ', min(date_totals, key=lambda x: x[1]))
    
overall_avg_consumption = viasatConsumption['Consumption'].sum() / len(date_groups)

print('\nOverall Average Consumption per day: ', overall_avg_consumption)
```

### Charging Data (Charge Duration, Energy Consumed, Gas Saved, etc.)

```python
# data cleaning 
viasatCharging_df_raw = pd.read_csv('data/viasatChargingData.csv')

# dropping columns with all null values, unnecessary features
viasatCharging_df_raw = viasatCharging_df_raw.drop(columns=[
    'Address 2', 
    'Fee', 
    'Start SOC', 
    'OnRamp ID Tag', 
    'Payment Terminal Session ID',
    'Org Name', 
    'Start Time Zone', 
    'End Time Zone', 
    'Currency', 
    'County',
    'End SOC',
    'CompanyID'])

# separating start date and end date into date and time columns
start_dates = viasatCharging_df_raw['Start Date'].apply(lambda x: datetime.strptime(x, '%m/%d/%Y %H:%M').date())
start_times = viasatCharging_df_raw['Start Date'].apply(lambda x: datetime.strptime(x, '%m/%d/%Y %H:%M').time())
end_dates = viasatCharging_df_raw['End Date'].apply(lambda x: datetime.strptime(x, '%m/%d/%Y %H:%M').date())
end_times = viasatCharging_df_raw['End Date'].apply(lambda x: datetime.strptime(x, '%m/%d/%Y %H:%M').time())

viasatCharging_df_raw['Start Date'], viasatCharging_df_raw['Start Time'] = start_dates, start_times
viasatCharging_df_raw['End Date'], viasatCharging_df_raw['End Time'] = end_dates, end_times

viasatCharging_df = viasatCharging_df_raw.copy()
viasatCharging_df.head()
```
